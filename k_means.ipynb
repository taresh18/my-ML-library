```
#  importing the required libraries
import k_means as KM
import numpy as np
import pandas as pd
```

```
# reading dataset
data = pd.read_csv("/content/linearreg.csv")
data.head()
```

```
# loading dataset from dataframes to numpy arrays
X = np.array(data)
X = X[:, 1:3].reshape(X.shape[0], 2)
X_train = X[:180, :].reshape(180, 2)
X_test = X[180:, :].reshape(20, 2)
```

```
# implementing K-means
a = KM.K_Means()
a.k_means(X_train, X_test, 3) # pass the value of k
```
